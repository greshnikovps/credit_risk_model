```python
import os
import pandas as pd
import tqdm


def read_parquet_dataset_from_local(path_to_dataset: str, start_from: int = 0,
                                     num_parts_to_read: int = 2, columns=None, verbose=False) -> pd.DataFrame:
    """
    читает num_parts_to_read партиций, преобразовывает их к pd.DataFrame и возвращает
    :param path_to_dataset: путь до директории с партициями
    :param start_from: номер партиции, с которой нужно начать чтение
    :param num_parts_to_read: количество партиций, которые требуется прочитать
    :param columns: список колонок, которые нужно прочитать из партиции
    :return: pd.DataFrame
    """

    res = []
    dataset_paths = sorted([os.path.join(path_to_dataset, filename) for filename in os.listdir(path_to_dataset)
                              if filename.startswith('train')])
    print(dataset_paths)

    start_from = max(0, start_from)
    chunks = dataset_paths[start_from: start_from + num_parts_to_read]
    if verbose:
        print('Reading chunks:\n')
        for chunk in chunks:
            print(chunk)
    for chunk_path in tqdm.tqdm_notebook(chunks, desc="Reading dataset with pandas"):
        print('chunk_path', chunk_path)
        chunk = pd.read_parquet(chunk_path,columns=columns)
        res.append(chunk)

    return pd.concat(res).reset_index(drop=True)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import OneHotEncoder
import numpy as np
    
class GroupById(BaseEstimator, TransformerMixin):
    def __init__(self, id_column):
        self.id_column = id_column
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.groupby(self.id_column).mean().reset_index(drop=True)

class OneHotEncoderWithDF(BaseEstimator, TransformerMixin):
    def __init__(self, **kwargs):
        self.encoder = OneHotEncoder(sparse_output=False, **kwargs)
        self.feature_names = None
    
    def fit(self, X, y=None):
        self.encoder.fit(X, y)
        self.feature_names = self.encoder.get_feature_names_out(X.columns)
        return self
    
    def transform(self, X):
        X_transformed = self.encoder.transform(X)
        transformed_df = pd.DataFrame(X_transformed, columns=self.feature_names)
        print(transformed_df)
        return transformed_df
    
class ColumnTransformerWithDF(BaseEstimator, TransformerMixin):
    def __init__(self, transformers):
        self.column_transformer = ColumnTransformer(transformers)
        self.feature_names = []
    
    def fit(self, X, y=None):
        self.column_transformer.fit(X, y)
        
        for name, trans, columns in self.column_transformer.transformers:
            if hasattr(trans, 'get_feature_names_out'):
                self.feature_names.extend(trans.get_feature_names_out(columns))
            else:
                self.feature_names.extend(columns)
                
        return self
    
    def transform(self, X):
        X_transformed = self.column_transformer.transform(X)
        return pd.DataFrame(X_transformed, columns=self.feature_names)
```

```python
def preprocess_dummy(df):
    return df
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from feature_engine.selection import DropFeatures

def add_new_features(df):
    df['total_overdue'] = (df['pre_loans5'] +
                                     df['pre_loans530'] +
                                     df['pre_loans3060'] +
                                     df['pre_loans6090'] +
                                     df['pre_loans90'])

    df['avg_days_since_open'] = (df['pre_since_opened'] + df['pre_since_confirmed']) / 2

    df['total_zero_loans'] = (df['is_zero_loans5'] +
                                        df['is_zero_loans530'] +
                                        df['is_zero_loans3060'] +
                                        df['is_zero_loans6090'] +
                                        df['is_zero_loans90'])
    
    df['max_next_pay_summ'] = df['pre_loans_next_pay_summ'].max()
    df['min_next_pay_summ'] = df['pre_loans_next_pay_summ'].min()

    df['diff_till_close'] = df['pre_till_fclose'] - df['pre_till_pclose']

    df['outstanding_to_credit_limit'] = df['pre_loans_outstanding'] / (df['pre_loans_credit_limit'] + 0.0000001)

    df['max_overdue_to_credit_limit'] = df['pre_loans_max_overdue_sum'] / (df['pre_loans_credit_limit'] + 0.0000001)

    # Создание бинарных признаков
    for i in range(36):
        if f'enc_paym_{i}' in df.columns:
            df[f'enc_paym_{i}_flag'] = df[f'enc_paym_{i}'].apply(lambda x: 1 if x > 0 else 0)
            
    return df

def replace_rare_categories(df, column, threshold=0.05):
    # заменяет редко встречающиеся категории числом -1
    freq = df[column].value_counts(normalize=True)
    rare_categories = freq[freq < threshold].index
    
    rare_categories_dict = {cat: -1 for cat in rare_categories}
    df[column] = df[column].replace(rare_categories_dict)

    return df

def preprocess_row_data(df):
    high_corr_to_remove = ['enc_paym_17', 'enc_paym_20', 'pre_over2limit', 'enc_paym_18', 'enc_paym_5', 'enc_paym_23',
                           'enc_paym_9', 'enc_paym_16', 'enc_paym_15', 'enc_paym_22', 'enc_paym_14', 'enc_paym_8',
                           'enc_paym_7', 'enc_paym_19', 'enc_paym_13', 'enc_paym_6', 'enc_paym_21', 'enc_paym_11',
                           'enc_paym_12', 'pre_loans_total_overdue']
    df = df.drop(high_corr_to_remove, axis=1)    
    
    df = add_new_features(df)
    
    binary_features = [col for col in df.columns if df[col].nunique() == 2]
    categorical_features = [col for col in df.columns if col not in 
                            binary_features + ['id', 'rn']]
    
    df = pd.get_dummies(df, columns=categorical_features, drop_first=True, dtype=np.int8)
    
    df = df.groupby('id').mean().reset_index()
    all_cols = df.columns.values[1:]
    
    df[all_cols] = (df[all_cols] * 255).astype(np.uint8)
    
    return df
```

```python
def prepare_transactions_dataset(path_to_dataset: str, num_parts_to_preprocess_at_once: int = 1, num_parts_total: int=50,
                                 save_to_path=None, verbose: bool=False, dummy_preprocess=False):
    """
    возвращает готовый pd.DataFrame с признаками, на которых можно учить модель для целевой задачи
    path_to_dataset: str
        путь до датасета с партициями
    num_parts_to_preprocess_at_once: int
        количество партиций, которые будут одновременно держаться и обрабатываться в памяти
    num_parts_total: int
        общее количество партиций, которые нужно обработать
    save_to_path: str
        путь до папки, в которой будет сохранён каждый обработанный блок в .parquet-формате; если None, то не будет сохранён
    verbose: bool
        логирует каждую обрабатываемую часть данных
    """
    preprocessed_frames = []

    for step in tqdm.tqdm_notebook(range(0, num_parts_total, num_parts_to_preprocess_at_once),
                                   desc="Transforming transactions data"):
        transactions_frame = read_parquet_dataset_from_local(path_to_dataset, step, num_parts_to_preprocess_at_once,
                                                             verbose=verbose)

       #здесь должен быть препроцессинг данных
        if dummy_preprocess:
            transactions_frame = preprocess_dummy(transactions_frame)
        else:
            transactions_frame = preprocess_row_data(transactions_frame)

       #записываем подготовленные данные в файл
        if save_to_path:
            block_as_str = str(step)
            if len(block_as_str) == 1:
                block_as_str = '00' + block_as_str
            else:
                block_as_str = '0' + block_as_str
            transactions_frame.to_parquet(os.path.join(save_to_path, f'processed_chunk_{block_as_str}.parquet'))

        preprocessed_frames.append(transactions_frame)
    return pd.concat(preprocessed_frames)
```

```python
# среди признаков есть коррелирующие, удалим их
threshold = 0.85
corr_pairs = correlation_matrix.abs().unstack().sort_values(kind="quicksort")
high_corr = [(x, y) for x, y in corr_pairs.index if (x != y) and (corr_pairs[x, y] > threshold)]

for x, y in high_corr:
    if (y, x) in high_corr:
        high_corr.remove((x, y))

high_corr_to_remove = set()
for x, y in high_corr:
    high_corr_to_remove.add(x)

high_corr_to_remove = list(high_corr_to_remove) + ['pre_loans_total_overdue']
high_corr_to_remove
```

### Разделим на train и test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(['flag', 'id'], axis=1), data['flag'], test_size=.2)
```

```python
X_train.shape
```

### Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

lr = LogisticRegression(class_weight='balanced', max_iter=3000)

lr.fit(X_train, y_train)
```

```python
probs = lr.predict_proba(X_test)[:, -1]
roc_auc_score(y_test, probs)
```

```python
probs = lr.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, probs)
```

Качество предсказаний ниже необходимого. Попробуем более сложную модель - XGBoost.

### XGBoost

```python
negative_class_count = sum(y_train == 0)
positive_class_count = sum(y_train == 1)

# Вычисление scale_pos_weight
scale_pos_weight = negative_class_count / positive_class_count
```

```python
from xgboost import XGBClassifier

bst = XGBClassifier(scale_pos_weight=scale_pos_weight, max_depth=2, n_estimators=250)

bst.fit(X_train, y_train)
```

```python
preds = bst.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, preds)
```

```python
preds = bst.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, preds)
```

Видим, что roc_auc_score больше 0.75, что соответствует требованиям задачи. Попробуем дополнительно убрать малозначимые признаки и обучить модешь без них

```python
import matplotlib.pyplot as plt
import seaborn as sns

importance = np.abs(lr.coef_[0])

# Создание DataFrame для удобства визуализации
feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importance
})

# Сортировка признаков по важности
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Визуализация важности признаков
plt.figure(figsize=(10, 25))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance in Logistic Regression')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
threshold = 0.0001

# Фильтрация признаков по важности
features_to_keep = feature_importance_df[feature_importance_df['Importance'] >= threshold]['Feature']

# Обновленный DataFrame с важными признаками
X_filtered = X_train[features_to_keep]
```

```python
x_filtered_test = X_test[features_to_keep]
```

```python
bst_filtered = XGBClassifier(scale_pos_weight=scale_pos_weight, max_depth=2, n_estimators=350)

bst_filtered.fit(X_filtered, y_train)
```

```python
from sklearn.metrics import roc_auc_score

preds = bst_filtered.predict_proba(X_filtered)[:, 1]
roc_auc_score(y_train, preds)
```

```python
preds = bst_filtered.predict_proba(x_filtered_test)[:, 1]
roc_auc_score(y_test, preds)
```
