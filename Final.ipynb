```python
import pandas as pd
import pickle
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

from xgboost import XGBClassifier

import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
data = pd.read_csv('train_data/final_data.csv', dtype=np.uint8)
data
```

## Разделим на train и test

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['flag', 'id'], axis=1), data['flag'], test_size=.2)
```

```python
X_train.shape
```

## Обучим модели

### Логистическая регрессия

```python
lr = LogisticRegression(class_weight='balanced', max_iter=3000)

lr.fit(X_train, y_train)
```

```python
probs = lr.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, probs)
```

```python
probs = lr.predict_proba(X_test)[:, -1]
roc_auc_score(y_test, probs)
```

Качество предсказаний ниже необходимого. Попробуем более сложную модель - XGBoost.

### XGBoost

```python
negative_class_count = sum(y_train == 0)
positive_class_count = sum(y_train == 1)

# Вычисление scale_pos_weight
scale_pos_weight = negative_class_count / positive_class_count
```

```python
bst = XGBClassifier(scale_pos_weight=scale_pos_weight, max_depth=2, n_estimators=250)

bst.fit(X_train, y_train)
```

```python
preds = bst.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, preds)
```

```python
preds = bst.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, preds)
```

Качество обучения неплохое. Попробуем уменьшить размерность датасета и обучить модель 

```python
importance = np.abs(lr.coef_[0])

# Создание DataFrame для удобства визуализации
feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importance
})

# Сортировка признаков по важности
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Визуализация важности признаков
plt.figure(figsize=(10, 25))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance in Logistic Regression')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
threshold = 0.0001

# Фильтрация признаков по важности
features_to_keep = feature_importance_df[feature_importance_df['Importance'] >= threshold]['Feature']

# Обновленный DataFrame с важными признаками
X_filtered = X_train[features_to_keep]
```

```python
x_filtered_test = X_test[features_to_keep]
```

```python
bst_filtered = XGBClassifier(scale_pos_weight=scale_pos_weight, max_depth=2, n_estimators=350)

bst_filtered.fit(X_filtered, y_train)
```

```python
preds = bst_filtered.predict_proba(X_filtered)[:, 1]
roc_auc_score(y_train, preds)
```

```python
preds = bst_filtered.predict_proba(x_filtered_test)[:, 1]

roc_auc_score(y_test, preds)
```

Видим, что обученная модель неплохо справляется с предсказаниями

## Создание Пайплайна

```python
def select_features(X, features):
    return X[features]

feature_selector = FunctionTransformer(select_features, kw_args={'features': features_to_keep})

pipeline = Pipeline([
    ('select_features', feature_selector),
    ('classifier', XGBClassifier(scale_pos_weight=scale_pos_weight))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
preds = pipeline.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, preds)
```

Сохраним тестовые предсказания в файл

```python
pd.DataFrame(preds).to_csv("predictions.csv")
```

Видим, что пайплайн отлично справляется с задачей (roc_auc = 0.7516628943619447). Обучим его на всех данных и сохраним в файл

```python
pipeline.fit(data.drop(['flag', 'id'], axis=1), data['flag'])
```

```python
with open('pipeline.pickle', 'wb') as f:
    pickle.dump(pipeline, f)
```

Итого, был произведен анализ фич, наимее важные из них не были взяты в рассмотрение. Также мы попробовали несколько алгоритмов машинного обучения и остановили свой выбор на XGBoost, который показал наилучшие результаты.
